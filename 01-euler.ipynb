$y'(t) = f(t, y)$
<br>
$ùë¶(ùë° + \delta t) = ùë¶(ùë°) + \delta t f(y(t))$

1a) Write a Python script that implements the Euler method and solve the differential
equation:

$\frac{dy}{dt} = y$

```python
interval = 0.001

ts = np.arange(0, 10, interval)
ys = [1]
```

```python
def f(y_at_t):
    return y_at_t
```

```python
for i in range(1, len(ts)):
    t = ts[i]
    prev_y = ys[i-1]
    new_val = prev_y + (interval * f(prev_y))
    ys.append(new_val)
```

```python
def exp(t):
    return np.exp(t)
```

```python
plt.title("Euler method")
plt.plot(ts, ys)
plt.plot(ts, exp(ts))
plt.show()
```

1b) <br>
$my''(t) = ‚àícy'(t) - ky(t)$
<br>
$p(t) = y'(t)$
<br>
<br>
$my''(t) = -cp(t) - ky(t)$

$ùë¶'(ùë° + \delta t) = ùë¶'(ùë°) + \delta t y''(t)$

```python
ps = [0]
ys = [1]
ts = np.arange(0, 100, interval)

m = 1
c = 0.1
k = 1
```

```python
for i in range(1, len(ts)):
    y_prime_t = ps[i-1]
    y_t = ys[i-1]
    y_prime_prime_t = -(((c/m) * (y_prime_t)) + ((k/m) * y_t))
    y_prime_new = y_prime_t + (interval * y_prime_prime_t)
    y_new = y_t + (interval * y_prime_t)
    ps.append(y_prime_new)
    ys.append(y_new)
```

```python
plt.plot(ts, ys)
```
